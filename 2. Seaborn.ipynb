### Categorical Categorical Plot
* Heatmaps
* Clustermaps

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
x = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='mean')
x
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(x)
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(x, cbar = False)
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(x,cbar = False, linewidths=0.5)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(x,cbar = False, linewidths=0.5, annot=True)
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(x,cbar = False, linewidths=0.5, annot=True, fmt='d')
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(x,cbar = False, linewidths=0.5, annot=True, fmt='d', cmap='viridis')
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(x,cbar = False, linewidths=0.5, annot=True, fmt='d', cmap='summer')
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(x,cbar = False, linewidths=0.5, annot=True, fmt='d', cmap='winter')
```

## Clustermap

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x, metric='correlation', z_score=0, annot = True)
```

```python
sns.clustermap(x, metric='correlation', z_score=0, annot = True, row_cluster=False)
```

```python
sns.clustermap(x, metric='correlation', z_score=0, annot = True, col_cluster=False)
```
