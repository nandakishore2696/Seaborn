## Joint Plot
* Jointplot creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on seperate axes

we can use jopintplots in the following 4 manners
1. Scatter plots
2. Hexbin Plots
3. Kdeplots
4. Regplot

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

## Pair PLots
* Pairplots are used to plot multiple pairwise bivariate distributions in a dataset
* This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame.
* Bydefault, it also draws the univariate distribution of each variable on the diagonal axes

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

## Distplots

* Distplots are used to do univsriste analysis on a continuous/Numerical data
Dist plots combine 3 thhings in one graph:
1. Histogram
2. Kernel Density Estimation(KDE)
3. RugPlot 

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
titanic
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'], kde=False)
```

```python
sns.distplot(titanic['age'], kde=False,bins=10)
```

```python
sns.distplot(titanic['age'], kde=False, bins=20)
```

```python
sns.distplot(titanic['age'], hist=False)
```

```python
sns.distplot(titanic['age'], rug=True, hist=False, kde=False)
```

```python
sns.distplot(titanic['age'],rug=True)
```

```python
sns.distplot(titanic[titanic['survived'] == 0]['age'])
```

```python
sns.distplot(titanic[titanic['survived'] == 1]['age'])
sns.distplot(titanic[titanic['survived'] == 0]['age'])
```
