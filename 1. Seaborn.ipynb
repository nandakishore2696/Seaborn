# Relational / Statistical Plots
 Relational Plots are used to perform MultiVariate Analysis
 1. Scatter Plots
 2. Line Plots (Specifically used for timeseries)
 3. FacetGrids (Plotting multiple graphs side by side)

## Relplot Plots
* Although, scatter plot can be plotted for both numerical as well asm categorical data, it's mostly used to depict relationship between  numerical  data

Trere are 2 ways to draw scatterplot in seaborn
1. relplot [Figure Level Function]
2. scatterplot [Axes Level Function]

* As we will see, these visualizations can be quite helpful because they use simple and easily-understood representations of data even on complex data by mapping upto three additional variables using semantics of hue, size and styye

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data
```

* Plot Scatter plot between total bill and tip

```python
# kind: Literal['scatter', 'line'] = "scatter"
sns.relplot(x='total_bill',y='tip',data = data, kind='scatter')
```

```python
# kind='line' 
sns.relplot(x='total_bill',y='tip',data = data, kind='line')
```

```python
# Hue - categorical data
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker')
```

```python
# Style - categorical data
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker', style='sex')
```

```python
# size - Numerical data
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker', style='sex',size='size')
```

```python
# sizes - range of sizes
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker', style='sex',size='size', sizes=(15,200))
```

## Scatter Plot

```python
sns.scatterplot(data=data, x='total_bill', y='tip', hue='smoker')
```

### Categorical Scatter Plots
* X --> Categorical Plot
* Y --> Numerical Plot

```python
data.head()
```

```python
# Strip Plot
sns.catplot(data=data, x='day', y='tip', kind='strip')

#jitter
```

```python
# Strip Plot
#jitter - the noise added to visualize the data
sns.catplot(data=data, x='day', y='tip', kind='strip', jitter = 1)
```

```python
sns.catplot(data=data, x='day', y='tip', kind='strip', jitter = 0)
```

```python
# swarn
sns.catplot(data=data, x ='day',y='tip', kind='swarm')
```

```python
sns.catplot(data=data,x='day',y='tip',kind='swarm', hue='sex')
```

```python
sns.swarmplot(data=data,x='day',y='tip', hue='sex')
```

###  Categorical Distribution Plots
1. Box Plot
2. Violin Plot - method of plotting numeric data and can be considered a combination of the box plot with a kernel density plot 
* The unquestionable advantage of violin plot over box plot is that aside from showing the above mentioned statistics, it also shows the entire distribution of the data
* Violin plots are often used to compare the distribution of a given variable across some categories


```python
# Univariate Analysis
sns.boxplot(data['tip'])
```

```python
sns.boxplot(data['total_bill'])
```

```python
sns.catplot(data=data, x='day', y='total_bill', kind='box')
```

```python
# hue
sns.catplot(data=data, x='day', y='total_bill', kind='box', hue='sex')
```

#### Violin Plot

```python
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(data=data, x='day', y='total_bill', kind='violin')
```

```python
sns.catplot(data=data, x='day', y='total_bill', kind='violin', hue='sex')
```

```python
# split
sns.catplot(data=data, x='day', y='total_bill', kind='violin', hue='sex', split = True)
```

### Categorical Estimate Plots
* Pointplot
* Barplot
* Counterplot

```python
sns.catplot(data=data, x='smoker', y='tip',kind='bar')
```

```python
sns.catplot(data=data, x='smoker', y='total_bill',kind='bar')
```

```python
sns.catplot(data=data, x='smoker', y='total_bill',kind='bar', hue='sex')
```
